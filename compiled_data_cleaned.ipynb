```python
import pandas as pd
```

```python
#load compiled data into a dataframe
df = pd.read_csv('./Compiled_Trips_Data/compiled_data.csv')
```

```python
#shows the top rows of the table -> shows us what the table looks like. 
df.head()
```

```python
df.info()
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
'''For this problem we are not concerned with the station name or ID, we want to answer the question | How do Casual riders and Members Differ? |
We will load a new data frame with these factors removed from consideration and ONLY track ride_id, rideable_type, start/end station, start/end time (we will add this later), member casual
'''
narrow_df = df[['ride_id','rideable_type','member_casual','start_station_name','end_station_name']]
```

```python
narrow_df.head()
```

```python
#let's check to make sure columns and rows exist as they should
narrow_df.shape
```

```python
#create new variable that stores the narrow data incase we have to start over
expanded_narrow_df = narrow_df
```

```python
'''We have our desired rows and columns. Next we need to look at segmenting date and time. The date itself will help show which days are the most popular.
The time will show when bikes are typically demanded and for how '''

expanded_narrow_df['start_date'] = pd.to_datetime(df['started_at'])
expanded_narrow_df['end_date'] = pd.to_datetime(df['ended_at'])
```

```python
expanded_narrow_df.head()
```

```python
'''Our new dataframe is more or less in the shape that we want it. WE are tracking the use characteristics of the member and casual groups.
Considering this data spans 2020 - 2022 we can gain more insight into how the customers interact with the service.
Now we need to clean the data of blank or null values. While this will impact the total count of our data, with over 1 million rows,
there is statistical significance even at a 70% (or ~ 300,000 rows) attrition rate

The next step is to understand how much of the current dataframe is missing/NaN and summarize that. 
'''
expanded_narrow_df.isnull().sum()
```

```python
'''Conveniently, we're only missing data from ~12.5k rows. 
Relatively, this is about 1% of all data values.
Next we will look at the ride_ids and ensure they are a similar format over the entire frame.
'''
expanded_narrow_df['ride_id'].str.len().value_counts()
```

```python
#dropping any row (index) with a missing value. 
dropped_df = expanded_narrow_df.dropna(how='any')
```

```python
#looking at index count and head for new df. 
dropped_df.info()
dropped_df.head()
```

```python
dup_df = dropped_df.drop_duplicates(subset='ride_id')
```

```python
dup_df.head()
```

```python
indexed_df = dup_df.set_index('ride_id')
```

```python
indexed_df.info()
indexed_df.head()
```

```python
indexed_df.to_csv('compiled_data_clean.csv',',',encoding='utf-8')
```
